**Comparing Models**

```python
import matplotlib as mpl
mpl.rcParams['axes.labelsize'] = "x-large"
mpl.rcParams['axes.titlesize'] = 'x-large'
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['ytick.labelsize'] = 'x-large'
import json
ratios = {}
lr_ratios = {}
with open("/home/user/vesta_csvs/32-probes-256-iters-faithful-McCullough-predictions.json") as fp:
    traces = json.load(fp)
    for tr in traces:
        bench = tr["benchmark"]
        dt_prediction_energy = np.array(tr["dt"])
        lr_prediction_energy = np.array(tr["lr"])
        nn_prediction_energy = np.array(tr["nn"])
        real_energy = np.array(tr["real"])
        fig = plt.gcf()
        fig, ax = plt.subplots(figsize=(5, 3))
        #ax.plot(range(0, len(dt_prediction_energy)), dt_prediction_energy, label="XGBoost", linestyle="--", color="tab:blue")
        #ax.plot(range(0, len(lr_prediction_energy)), lr_prediction_energy, label="LinearRegression", linestyle="-.", color="tab:orange")
        #ax.plot(range(0, len(nn_prediction_energy)), nn_prediction_energy, label="Neural Network", linestyle=":", color="tab:green")
        ax.plot(range(0, len(real_energy)), real_energy, label="ground truth", color="black")
        ax.plot(range(0, len(dt_prediction_energy)), dt_prediction_energy, label="XGBoost", linestyle="--", color="tab:blue")
        dt_total_prediction_energy = dt_prediction_energy.sum()
        lr_total_prediction_energy = lr_prediction_energy.sum()
        nn_total_prediction_energy = nn_prediction_energy.sum()
        total_actual_energy = real_energy.sum()
        dt_ratio =  np.abs((total_actual_energy - dt_total_prediction_energy) / total_actual_energy)
        lr_ratio =  np.abs((total_actual_energy - lr_total_prediction_energy) / total_actual_energy)
        nn_ratio =  np.abs((total_actual_energy - nn_total_prediction_energy) / total_actual_energy)
        ratios[bench] = dt_ratio
        lr_ratios[bench] = lr_ratio
        print(f"{bench} -- dt: {dt_ratio * 100:.2f}%; lr: {lr_ratio * 100:.2f}%; nn: {nn_ratio * 100:.2f}%")
        #if bench == "avrora":
        #    plt.title(f"Prediction Power vs Real Power")
        plt.xlabel("Time (s)")
        plt.ylabel("Power (W)")
        plt.savefig(f"/home/user/vesta_figs/PLDI_25_256-iters_Predictions_XGBOOST-only/{bench}.pdf", format="pdf", bbox_inches="tight")
        plt.show()
        plt.clf()        
```

```python
import matplotlib as mpl
mpl.rcParams['axes.labelsize'] = "x-large"
mpl.rcParams['axes.titlesize'] = 'x-large'
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['ytick.labelsize'] = 'x-large'
means = pd.DataFrame(ratios, index=[0]).transpose().rename(columns={0:"mean"}).sort_index()
means_lr = pd.DataFrame(lr_ratios, index=[0]).transpose().rename(columns={0:"mean"}).sort_index()
random_samples = means.sample(n=37).index.sort_values()
N = len(means.loc[random_samples])
ind = np.arange(N)
width = .5
from matplotlib.ticker import PercentFormatter
import matplotlib.ticker as mtick
fig = plt.gcf()
fig, ax = plt.subplots(figsize=(10, 3))
#ax.bar(ind, means.loc[random_samples]["mean"]*100, width, capsize=2, color="tab:blue", edgecolor="black")
ax.bar(ind, means_lr.loc[random_samples]["mean"]*100, capsize=2, color="tab:orange", edgecolor="black",hatch="-")
#plt.title(f"Abs Error: VESTA XGBoost vs Linear Model")
plt.ylim(0,90)
plt.xlabel("Benchmark")
plt.ylabel("Percent Error")
fmt = '%.0f%%' # Format you want the ticks, e.g. '40%'
yticks = mtick.FormatStrFormatter(fmt)
ax.yaxis.set_major_formatter(yticks)
plt.xticks(ind, means.loc[random_samples].index)
ax.set_xlim(ind[0]-1, len(means))
plt.xticks(rotation=55, ha='right')
plt.savefig(f"/home/user/vesta_figs/prediction-error-linear-model.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.clf()
```

```python
import dtreeviz
import pandas as pd
from xgboost import XGBRegressor
import numpy as np
import matplotlib.pyplot as plt
import shap

df_train = pd.read_csv("/home/user/training_data_for_visualized_model.csv")
#df_train = df_train[df_train.benchmark == "sunflow"]
events_df_train = df_train
events_df_train = events_df_train.drop(["iteration","ts","benchmark","power","cpu_power","memory_power"], axis=1)
events_df_train = events_df_train.drop([col for col in events_df_train.columns if 'energy_component' in col], axis=1)
power_df_train = df_train["power"]
df_test = pd.read_csv("/home/user/testing_data_for_visualized_model.csv")

model = XGBRegressor(tree_method='gpu_hist', gpu_id=0)
model.load_model("/home/user/xgboost-visualizer.json")
explainer = shap.explainers.TreeExplainer(model, events_df_train)
shap_values = explainer(events_df_train)
shap_values = shap_values
```

```python
import matplotlib as mpl
mpl.rcParams['axes.labelsize'] = "x-large"
mpl.rcParams['axes.titlesize'] = 'x-large'
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['ytick.labelsize'] = 'x-large'
features = ["thread__park","SetIntField","SetByteArrayRegion",
            "NewStringUTF","gc","vmops","thread__sleep","GetStringLength"]
for f in features:
    fig, ax = plt.subplots(figsize=(10, 3))
    shaps = shap_values[:, [f]].values
    feature = shap_values[:, [f]].data
    s_arr = []
    f_arr = []
    p_arr = []
    for i in range(len(shaps)):
        if feature[i][0] == -1 or feature[i][0] == 0: 
                continue
        s_arr.append(shaps[i][0])
        f_arr.append(feature[i][0])
        if power_df_train[i] > 150:
            p_arr.append(150)
        else:
            p_arr.append(power_df_train[i])
    plot = ax.scatter(f_arr, s_arr, c=p_arr, cmap="bwr", alpha=.33)
    plt.title(f"{f} SHAP Values")
    plt.xlabel("Depth")
    plt.ylabel("SHAP value")
    fig.colorbar(plot, label="power")
    plt.axhline(y = 0.0, color = 'r', linestyle = '--') 
    ax.set_xlim(1)
    plt.xticks(rotation=15, ha='center')
    plt.savefig(f"/home/user/vesta_figs/SHAP_{f}_scatter.pdf", format="pdf", bbox_inches="tight")
    plt.show()
    plt.clf()
```

```python
#features = ["thread__park","SetIntField","SetByteArrayRegion",
#            "NewStringUTF","gc","vmops","thread__sleep","GetStringLength"]
features = ["GetStringLength", "thread__sleep", "vmops", "gc", "NewStringUTF", 
            "SetByteArrayRegion", "SetIntField", "thread__park"]
violin_plot_format = []
for f in features:
    shaps = shap_values[:, [f]].values.tolist()
    shaps_f = shap_values[:, [f]].data.tolist()
    arr = []
    for i in range(len(shaps)):
        if shaps_f[i][0] == -1 or shaps_f[i][0] == 0: 
                continue
        arr.append(shaps[i][0])
    violin_plot_format.append(arr)
```

```python
features = ["thread__park","SetIntField","SetByteArrayRegion","NewStringUTF","gc","vmops","thread__sleep",
            "GetStringLength","GetObjectClass","GetMethodID","method__compile","safepoint","GetObjectArrayElement",
            "CallVoidMethod","compiled__method__load","NewString","GetLongField","compiled__method__load",
            "NewString","IsInstanceOf","GetEnv","CallObjectMethod","ReleaseIntArrayElements",
            "GetByteArrayElements","Throw","compiled__method__unload"]
features.reverse()

importance = []
stds = []
for f in features:
    shaps = shap_values[:, [f]].values.tolist()
    importance.append(np.absolute(shaps).mean())
    stds.append(np.absolute(shaps).std())
```

```python
import matplotlib as mpl
mpl.rcParams['axes.labelsize'] = "x-large"
mpl.rcParams['axes.titlesize'] = 'x-large'
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['ytick.labelsize'] = 'x-large'
from matplotlib.ticker import PercentFormatter
import matplotlib.ticker as mtick

fig = plt.gcf()
fig, ax = plt.subplots(figsize=(10, 3))
ax.bar(features, importance, yerr=stds,capsize=2, color="tab:pink", edgecolor="black")
#plt.title(f"Feature Importance")
plt.ylim(0,18)
plt.xlabel("Runtime Event")
plt.ylabel("Avg Abs SHAP Value")
#ax.set_xlim(0.5, len(features))
plt.xticks(rotation=55, ha='right')
plt.savefig(f"/home/user/vesta_figs/SHAP-ranked-importance.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.clf()
```
